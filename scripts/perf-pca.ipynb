```python
# loading the large language model kernels
for file in glob("./ncu-reports/*.ncu-rep"):
    os.system(f"ncu --import {file} --csv >> {file}.csv")
    print(file)
```

```python
dfs = []

for file in glob("*.csv"):
    df = pd.read_csv(file)
    df['source'] = 'rodinia'
    df['cmd'] = file.replace("ncu-reports/", "").replace(".csv", "")
    base_name = os.path.splitext(file)[0]
    if len(df.columns) < 5:
        continue
    dfs.append(df)

for file in glob("ncu-reports/*.csv"):
    df = pd.read_csv(file)
    df['source'] = 'huggingface'
    df['cmd'] = file.replace("ncu-reports/", "").replace(".csv", "").replace(".ncu-rep", "").upper()
    base_name = os.path.splitext(file)[0]
    if len(df.columns) < 5:
        continue
    dfs.append(df)

# Concatenate all DataFrames
dfs = pd.concat(dfs, axis=0, ignore_index=True)
dfs = dfs.drop_duplicates(subset=["Kernel Name", "Metric Name"])
dfs['Metric Value'] = dfs['Metric Value'].apply(lambda x: str(x).replace(",", "") if isinstance(x, str) else x)
dfs
```

```python
# pivoting the data
df = dfs.pivot(index=["cmd", "source", 'Kernel Name'], columns='Metric Name', values='Metric Value')

# df.columns = [column for column in df.columns]
# df['gpc__cycles_elapsed.max'] = df['gpc__cycles_elapsed.max'].apply(lambda x: int(str(x).replace(",", "")))
# df.reset_index(inplace=True)

for column in df.columns:
    # print(column)
    # if "avg.pct_of_peak" in column:
    #     del df[column]
    #     continue
    try:
        df[column] = df[column].astype(float)
    except:
        del df[column]

df.transpose()
```

```python
# Scaling and performing PCA
scaler = StandardScaler()
pca = PCA(n_components=2)

scaled_data = scaler.fit_transform(df)
principal_components = pca.fit_transform(scaled_data)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'], index=df.index)
pca_df = pca_df.reset_index()
pca_df
```

```python
import seaborn as sns

sns.scatterplot(pca_df, x='PC1', y='PC2', hue='source', alpha=0.3)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Rodinia Benchmarks vs The Top Trending Models\n on Hugging Face on the Week of 8/26')
plt.savefig("pca_of_rodinia.png")
plt.show()
```
